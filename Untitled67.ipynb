```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
from tqdm import tqdm
import matplotlib.animation as animation
from IPython.display import HTML, display

"""
Optimized N-body Simulation based on Philip Mocz (2020)
Princeton University, @PMocz
With GIF export capability for Google Colab
"""

@jit(nopython=True)
def calculate_acceleration(pos, mass, G, softening):
    """
    Calculate the acceleration on each particle due to Newton's Law of Gravity
    Uses Numba for significant performance improvement
    """
    N = pos.shape[0]
    acc = np.zeros((N, 3))

    # Compute accelerations
    for i in range(N):
        for j in range(N):
            if i != j:
                dx = pos[j, 0] - pos[i, 0]
                dy = pos[j, 1] - pos[i, 1]
                dz = pos[j, 2] - pos[i, 2]

                inv_r3 = (dx*dx + dy*dy + dz*dz + softening*softening)**(-1.5)

                acc[i, 0] += G * mass[j] * dx * inv_r3
                acc[i, 1] += G * mass[j] * dy * inv_r3
                acc[i, 2] += G * mass[j] * dz * inv_r3

    return acc

def calculate_energy(pos, vel, mass, G):
    """
    Calculate kinetic and potential energy of the system
    """
    # Kinetic Energy
    KE = 0.5 * np.sum(mass * np.sum(vel**2, axis=1))

    # Potential Energy
    PE = 0.0
    N = pos.shape[0]

    for i in range(N):
        for j in range(i+1, N):  # Sum over upper triangle
            dx = pos[j, 0] - pos[i, 0]
            dy = pos[j, 1] - pos[i, 1]
            dz = pos[j, 2] - pos[i, 2]

            inv_r = 1.0 / np.sqrt(dx*dx + dy*dy + dz*dz)
            PE -= G * mass[i] * mass[j] * inv_r

    return KE, PE

def leapfrog_step(pos, vel, acc, dt, mass, G, softening):
    """
    Execute one leapfrog integration step
    """
    # Half-kick
    vel += acc * dt/2.0

    # Drift
    pos += vel * dt

    # Update accelerations
    acc = calculate_acceleration(pos, mass, G, softening)

    # Half-kick
    vel += acc * dt/2.0

    return pos, vel, acc

def run_simulation(N=100, tEnd=10.0, dt=0.01, softening=0.1, G=1.0, frames_to_save=100):
    """
    Run the N-body simulation with specified parameters

    Parameters:
    -----------
    N : int
        Number of particles
    tEnd : float
        End time for simulation
    dt : float
        Time step
    softening : float
        Softening parameter to avoid numerical instability
    G : float
        Gravitational constant
    frames_to_save : int
        Number of frames to save for the animation

    Returns:
    --------
    simulation_data : dict
        Dictionary containing positions, times, and energies
    """
    # Initialize simulation
    np.random.seed(17)

    mass = 20.0 * np.ones(N) / N  # Total mass of particles is 20
    pos = np.random.randn(N, 3)   # Random positions and velocities
    vel = np.random.randn(N, 3)

    # Convert to Center-of-Mass frame
    vel -= np.sum(mass.reshape(-1, 1) * vel, axis=0) / np.sum(mass)

    # Calculate initial acceleration
    acc = calculate_acceleration(pos, mass, G, softening)

    # Calculate initial energy
    KE, PE = calculate_energy(pos, vel, mass, G)

    # Setup for time evolution
    Nt = int(np.ceil(tEnd/dt))
    save_interval = max(1, Nt // frames_to_save)

    # Save energies and positions
    frames_count = min(frames_to_save, Nt+1)
    pos_save = np.zeros((N, 3, frames_count))
    times_save = np.zeros(frames_count)
    KE_save = np.zeros(frames_count)
    PE_save = np.zeros(frames_count)

    # Save initial state
    pos_save[:, :, 0] = pos
    times_save[0] = 0
    KE_save[0] = KE
    PE_save[0] = PE

    # Main simulation loop
    t = 0
    saved_frame_idx = 1

    # Use tqdm for progress bar (works in Colab)
    for i in tqdm(range(Nt), desc="Running simulation"):
        # Leapfrog integration step
        pos, vel, acc = leapfrog_step(pos, vel, acc, dt, mass, G, softening)

        # Update time
        t += dt

        # Save data at specified intervals
        if i % save_interval == 0 and saved_frame_idx < frames_count:
            # Calculate energy
            KE, PE = calculate_energy(pos, vel, mass, G)

            # Save position and energy data
            pos_save[:, :, saved_frame_idx] = pos
            times_save[saved_frame_idx] = t
            KE_save[saved_frame_idx] = KE
            PE_save[saved_frame_idx] = PE

            saved_frame_idx += 1

    # Package simulation data
    simulation_data = {
        'positions': pos_save,
        'times': times_save,
        'KE': KE_save,
        'PE': PE_save,
        'total_energy': KE_save + PE_save
    }

    return simulation_data

def create_animation(simulation_data, filename='nbody.gif', fps=15, dpi=100):
    """
    Create and save an animation from the simulation data

    Parameters:
    -----------
    simulation_data : dict
        Dictionary containing positions, times, and energies
    filename : str
        Name of the output GIF file
    fps : int
        Frames per second for the animation
    dpi : int
        Dots per inch for the animation

    Returns:
    --------
    anim : matplotlib.animation.FuncAnimation
        The animation object
    """
    # Extract data
    pos_save = simulation_data['positions']
    times = simulation_data['times']
    KE = simulation_data['KE']
    PE = simulation_data['PE']
    total_energy = simulation_data['total_energy']

    # Create figure and axes
    fig = plt.figure(figsize=(8, 10), dpi=dpi)
    grid = plt.GridSpec(3, 1, wspace=0.0, hspace=0.3)
    ax1 = plt.subplot(grid[0:2, 0])
    ax2 = plt.subplot(grid[2, 0])

    # Set up axis limits and labels
    ax1.set(xlim=(-2, 2), ylim=(-2, 2))
    ax1.set_aspect('equal', 'box')
    ax1.set_xticks([-2, -1, 0, 1, 2])
    ax1.set_yticks([-2, -1, 0, 1, 2])

    ax2.set(xlim=(times[0], times[-1]), ylim=(-300, 300))
    ax2.set_xlabel('time')
    ax2.set_ylabel('energy')

    # Initial plots
    particles_plot = ax1.scatter([], [], s=10, color='blue')
    trails_plot = ax1.scatter([], [], s=1, color=[.7, .7, 1])

    ke_line, = ax2.plot([], [], 'r-', lw=1, label='KE')
    pe_line, = ax2.plot([], [], 'b-', lw=1, label='PE')
    etot_line, = ax2.plot([], [], 'k-', lw=1, label='Etot')
    ax2.legend(loc='upper right')

    # Title with time
    title = ax1.set_title("")

    # Number of frames
    n_frames = pos_save.shape[2]

    # Animation update function
    def update(frame):
        # Get current positions
        pos = pos_save[:, :, frame]

        # Get trail data (past 50 frames or less)
        trail_start = max(0, frame - 50)
        xx = pos_save[:, 0, trail_start:frame+1]
        yy = pos_save[:, 1, trail_start:frame+1]

        # Flatten for scatter plot
        x_trail = xx.flatten()
        y_trail = yy.flatten()

        # Update plots
        particles_plot.set_offsets(np.column_stack([pos[:, 0], pos[:, 1]]))
        trails_plot.set_offsets(np.column_stack([x_trail, y_trail]))

        # Update energy plots up to current frame
        ke_line.set_data(times[:frame+1], KE[:frame+1])
        pe_line.set_data(times[:frame+1], PE[:frame+1])
        etot_line.set_data(times[:frame+1], total_energy[:frame+1])

        # Update title with current time
        title.set_text(f"Time: {times[frame]:.2f}")

        return particles_plot, trails_plot, ke_line, pe_line, etot_line, title

    # Create animation
    anim = animation.FuncAnimation(
        fig, update, frames=n_frames,
        interval=1000/fps, blit=True
    )

    # Save as GIF
    anim.save(filename, writer='pillow', fps=fps, dpi=dpi)

    plt.close(fig)

    return anim

def run_and_animate(N=100, tEnd=10.0, dt=0.01, frames=100, filename='nbody.gif', display_in_notebook=True):
    """
    Run simulation and create animation in one step
    """
    # Install required packages if needed (for Colab)
    try:
        import google.colab
        IN_COLAB = True
        print("Running in Google Colab. Installing required packages if needed...")
        !pip install -q numba tqdm matplotlib
    except:
        IN_COLAB = False

    # Run simulation
    simulation_data = run_simulation(
        N=N,
        tEnd=tEnd,
        dt=dt,
        frames_to_save=frames
    )

    # Create animation
    anim = create_animation(simulation_data, filename=filename)

    # Display in notebook if requested
    if display_in_notebook and IN_COLAB:
        # Display the saved GIF file in the notebook
        from IPython.display import Image
        display(Image(filename))

        # Also create a simple plot to view final state
        plt.figure(figsize=(10, 5))

        # Final positions
        plt.subplot(1, 2, 1)
        final_pos = simulation_data['positions'][:, :, -1]
        plt.scatter(final_pos[:, 0], final_pos[:, 1], s=10, color='blue')
        plt.xlim(-2, 2)
        plt.ylim(-2, 2)
        plt.title("Final Positions")
        plt.aspect('equal')

        # Energy plot
        plt.subplot(1, 2, 2)
        plt.plot(simulation_data['times'], simulation_data['KE'], 'r-', label='KE')
        plt.plot(simulation_data['times'], simulation_data['PE'], 'b-', label='PE')
        plt.plot(simulation_data['times'], simulation_data['total_energy'], 'k-', label='Etot')
        plt.xlabel('Time')
        plt.ylabel('Energy')
        plt.legend()
        plt.title("Energy Conservation")

        plt.tight_layout()
        plt.show()

    return simulation_data
```
